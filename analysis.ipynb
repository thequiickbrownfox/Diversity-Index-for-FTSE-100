```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_excel('FTSE100 Index _ Data.xlsx', sheet_name='diversity_indicators')
df.head()
```

```python
print (df.shape)
print (df.columns.tolist())
```

```python
df= df.replace('MISSING', np.nan)
df.head()
```

```python
id_cols= ['company_name', 'ticker']
indicator_cols= [col for col in df.columns if col not in id_cols]
df[indicator_cols]= df[indicator_cols].apply(pd.to_numeric, errors='coerce')
df.head()
```

```python
df[indicator_cols]= df[indicator_cols].fillna(0)
df.head()
```

```python
##For Normalising the data, we are using the Min-Max Scaling technique.##
```

```python
scaled_df= df.copy()
for col in indicator_cols:
    min_value= scaled_df[col].min()
    max_value= scaled_df[col].max()
    if max_value - min_value != 0:
        scaled_df[col]= (scaled_df[col] - min_value) / (max_value - min_value)
    else:
        scaled_df[col]= 0.0
```

```python
reverse_scaled_cols= ['equality_gender_pay','equality_gender_pay_previous',
                      'equality_bame_pay','equality_lgbt_pay',
                      'equality_disability_pay','equality_gender_bonus']
print (reverse_scaled_cols)
```

```python
for col in reverse_scaled_cols:
    if col in scaled_df.columns:
        scaled_df[col]= 1 - scaled_df[col]
```

```python
scaled_df['diversity_score']= scaled_df[indicator_cols].mean(axis=1)
```

```python
scaled_df['rank']= scaled_df['diversity_score'].rank(ascending=False, method='min').astype(int)
scaled_df= scaled_df.sort_values(by='rank')
scaled_df.tail(20)
```

```python
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
```

```python
top_10= scaled_df.sort_values(by='diversity_score', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='company_name', y='diversity_score', data=top_10, palette='viridis')
plt.xticks(rotation=45, ha='right') 
plt.title("Top 10 FTSE 100 Companies by Diversity Commitment")
plt.xlabel("Company")
plt.ylabel("Diversity Score")

plt.show()
```

```python
bottom_10= scaled_df.sort_values(by='diversity_score').head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='company_name', y='diversity_score', data=bottom_10, palette='Reds')
plt.xticks(rotation=45, ha='right')
plt.title("Lowest Ranked FTSE 100 Companies by Diversity Commitment")
plt.xlabel("Company")
plt.ylabel("Diversity Score")

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(scaled_df['diversity_score'], bins=20, kde=True, color='blue')
plt.title("Distribution of Diversity Scores among FTSE 100 Companies")
plt.xlabel("Diversity Score")
plt.ylabel("Frequency") 
plt.show()
```

```python
for col in indicator_cols:
    idx = df[col].idxmax()
    company = df.loc[idx, 'company_name']
    value = df.loc[idx, col]
    print(f"{col} → highest: {company} ({value})")
```

```python
for col in indicator_cols:
    idx = df[col].idxmin()
    company = df.loc[idx, 'company_name']
    value = df.loc[idx, col]
    print(f"{col} → lowest: {company} ({value})")
```
